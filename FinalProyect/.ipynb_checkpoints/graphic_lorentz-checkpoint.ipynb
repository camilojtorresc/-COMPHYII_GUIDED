```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Para escribir los datos en arrays diferentes 
with open('data_lorentz.txt', 'r') as file:
    data = file.read().split('\n')

t = np.array([])
X = np.array([])
Y = np.array([])
Z = np.array([])

for i in range (0,len(data)-1):
    arr_string = data[i].split(";")
    
    t = np.append(t, float(arr_string[0]))
    X = np.append(X, float(arr_string[1]))
    Y = np.append(Y, float(arr_string[2]))
    Z = np.append(Z, float(arr_string[3]))
```

```python
sigma = 10.0; beta = 8/3; alpha = np.sqrt((4/beta) - 1);ro = 28.0;kappa = 0.35;

dxdt = -sigma*X+sigma*Y;
dydt = -X*Z+ro*X-Y; 
dzdt =  X*Y-beta*Z;
```

```python
f, axarr = plt.subplots(3, sharex=True,figsize=(10, 7))
axarr[0].plot(t, X)
axarr[0].set_title('LORENTZ SYSTEM SOLUTIONS')
axarr[0].grid()
axarr[0].set_ylabel("X")
axarr[1].plot(t, Y,"r")
axarr[1].grid()
axarr[1].set_ylabel("Y")
axarr[2].plot(t, Z,"g")
axarr[2].set_ylabel("Z")
axarr[2].grid()
axarr[2].set_xlabel("t")

plt.show()
```

```python
f, axarr = plt.subplots(3, sharex=True,figsize=(10, 7))
axarr[0].plot(t, dxdt)
axarr[0].set_title('LORENTZ SYSTEM DERIVATIVE SOLUTIONS')
axarr[0].grid()
axarr[0].set_ylabel("dxdt")
axarr[1].plot(t, dydt,"r")
axarr[1].grid()
axarr[1].set_ylabel("dydt")
axarr[2].plot(t, dzdt,"g")
axarr[2].set_ylabel("dzdt")
axarr[2].grid()
axarr[2].set_xlabel("t")

plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 7))
fig.suptitle('PHASE SPACES')

ax1.plot(X, dxdt)
ax1.set_xlabel("X")
ax1.set_ylabel("dxdt")
ax1.set_title("Phase space for X")
ax1.grid()

ax2.plot(Y, dydt,"r")
ax2.set_xlabel("Y")
ax2.set_ylabel("dydt")
ax2.set_title("Phase space for Y")
ax2.grid()

ax3.plot(Z, dzdt, "g")
ax3.set_xlabel("Z")
ax3.set_ylabel("dzdt")
ax3.set_title("Phase space for Z")
ax3.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(15, 10))
ax = fig.gca(projection='3d')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D PHASE SPACE')
ax.plot(X, Y, Z, label='rk4 method')
ax.legend()
ax.grid()

plt.show()
```
